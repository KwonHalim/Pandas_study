## Data Visualization with Matplotlib

### The plot() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots= True, figsize=(15, 12), sharex= False, sharey=False)
plt.show()
# subplot은 각각의 열로 분리된 플롯을 볼 수 있게 해준다.
# figsize는 그래프의 높이와 너비를 설정한다.
# x축, y축이 각각의 그래프에 나타나게 해주는 것이 sharex, sharey
```

```python
titanic.age.plot(figsize=(12, 8))
plt.show()
```

### Customization of plots

```python
xticks = [x for x in range(0,901, 50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
plt.style.available
```

```python
plt.style.use("classic")
```

```python
titanic.age.plot(figsize = (12,8), fontsize= 13, c = "r", linestyle = "-", # c 는 색깔을 뜻함
                 xlim = (0,len(titanic)), ylim = (titanic.age.min(),titanic.age.max())#xlim과 ylim은 각각 볼 수 있는 자료의 크기를 정함 
                 , xticks = xticks,yticks= yticks, rot=45)# xticks와 yticks는 위에 정의, rot는 글자 회전
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = 2, fontsize = 15) 
#loc를 통해 그래프 정보를 나타내는 창의 위치 변경 가능 ex)오른쪽위 ,왼쪽아래, 왼쪽위 등등
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid() #뒤의 격자 나타내기
plt.show()
```

```python
plt.style.available
#기본적으로 지원하는 plt 스타일들 확인하기
plt.style.use("seaborn-v0_8")
#특정 스타일 사용하는 방법
```

### Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins=15)
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins = 80, density = True)
# Shift + tab으로 나타나지 않는 것도 있음, 예를 들어 bins->나오지 않지만 히스토그램에서 사용 가능
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize=15, ylabelsize= 15, cumulative = False)
# 위에서는 kind='hist'로 했지만, 여기서는 .hist로 직접 사용
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative= True)
# hist 호출시 , NaN값들을 제거하고 호출해야함 ->X 매개변수 자리
# density=True 상대 개수로 표시, 그래서 cumulative = True일 때 마지막 값이 1로 정규화된 수치 확인 가능한 것
plt.show()
```

### Barcharts and Piecharts

### Scatterplots

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x = "age", y = "fare", c = "pclass"
             , marker = "X", s = 80, colormap= "viridis" )
#marker는 점의 표시 방법을 바꿈, s는 점의 크기를 바꿈
# c는 원래 색깔이지만 ,다른 Column을 추가해주면 3차원을 표시하는 데이터로 변환 가능
#colormap은 matplotlib에서 보고 색깔보고 선택하면 된다고 함
plt.show()
```
