### Customization of plots

```python
xticks = [x for x in range(0,901, 50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
plt.style.available
```

```python
plt.style.use("classic")
```

```python
titanic.age.plot(figsize = (12,8), fontsize= 13, c = "r", linestyle = "-", # c   
                 xlim = (0,len(titanic)), ylim = (titanic.age.min(),titanic.age.max())#xlim ylim        
                 , xticks = xticks,yticks= yticks, rot=45)# xticks yticks  , rot  
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = 2, fontsize = 15) 
#loc         ex) ,,  
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid() #  
plt.show()
```

### Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins=15)
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins = 80, density = True)
# Shift + tab    ,   bins->    
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize=15, ylabelsize= 15, cumulative = False)
#  kind='hist' ,  .hist  
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative= True)
# hist  , NaN   ->X  
# density=True   ,  cumulative = True    1     
plt.show()
```

### Barcharts and Piecharts

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
summer_2012 = pd.read_csv("summer_2012.csv", index_col = "Country")
```

```python
summer_2012
```

```python
summer_2012.Medal.plot(kind = "bar", figsize = (12,8), fontsize = 12, ylim=(0,1000))
plt.show()
```

```python
summer_2012.Medal.plot(kind = "pie", figsize = (12,8), fontsize = 12, autopct='%1.1f%%')
plt.show()
```
