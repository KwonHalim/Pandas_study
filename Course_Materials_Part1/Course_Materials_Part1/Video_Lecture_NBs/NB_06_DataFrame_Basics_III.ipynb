### Ranking DataFrames with rank()

```python
import pandas as pd
```

```python
sales = pd.Series([15, 32, 45, 21, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales = pd.Series([15, 32, 45, 15, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank(ascending=False, method = "min").sort_values(ascending = True)
#등수 매기기, method는 등수가 정수로 정렬되게 해줌, method가 없으면 값의 평균으로 예를 들어 5등이 2명이면
#2명의 등수가 5.5 가 된다. 기본값이 average이기 때문에 평균값으로 넣는 것
```

```python
sales.rank(ascending=False, method="min", pct=True).sort_values()
```

```python
distribution_percent = (sales / sales.sum()) * 100
print(distribution_percent)
#각 데이터 분포퍼센트로 나타내기
```

### The agg() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.mean()
```

```python
titanic.agg("mean", numeric_only=True)
```

```python
#titanic.agg(["mean", "std"]) 원래 이건데 버전이 달라져서 그런가 오류가 뜸
#그래서 dtype이 jnt형식일때만 agg할 수 있도록 수정해봄
#titanic.info()
titanic_int_float = titanic.select_dtypes(include=['int', 'float'])
titanic_int_float.agg(["mean", "std"])
```

```python
#titanic.agg(["mean", "std", "min", "max", "median"])
#얘도 마찬가지로 오류가 뜬다.
titanic_int_float = titanic.select_dtypes(include=['int', 'float'])
titanic_int_float.agg(["mean", "std", "min", "max", "median"])
```

```python
titanic.agg({"survived": "mean", "age":["min", "max"]})
#survived는 평균값만 보고 age는 최소값 및 최대값만 확인할 때
#리스트 대신 딕셔너리로 이용하면 된다.
```

```python
titanic.describe()
#describe랑 다른점? 
#agg는 사용자 지정 집계 함수 적용 가능
#dsscribe는 기본 통계 요약 제공
```

```python
def range(series):
    return series.max() - series.min()
```
